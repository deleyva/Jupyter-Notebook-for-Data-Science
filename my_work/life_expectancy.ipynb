```python
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/life_expectancy/API_SP.DYN.LE00.IN_DS2_en_csv_v2_10181296.csv', header=2, index_col=(0, 1, 2, 3))
df.head()
```

```python
life_expectancy = df.groupby(level=0).first().transpose()
life_expectancy.head()
```

```python
life_expectancy_subset = life_expectancy[['China', 'United States', 'Germany', 'Spain']]
```

```python
life_expectancy_subset.plot()
```
